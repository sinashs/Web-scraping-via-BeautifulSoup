```python
import socket
from urllib import request, parse, error
from urllib.request import urlopen
import re
from bs4 import BeautifulSoup
import ssl
import pandas as pd
```

```python
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

```python
html = urllib.request.urlopen('https://feinternational.com/buy-a-website/', context = ctx).read()
soup = BeautifulSoup(html, 'html.parser')
```

```python
link = []
tags = soup('a')
for tag in tags:
    #print(tag.get('href', None))
    link.append(tag.get('href', None))
```

```python
""" Collect the description of each busuiness"""
description = []
word_list = []
Industry = []
Category = []
Summary = []

tags = soup('h2')
for tag in tags:
    line = tag.contents[0]
    for word in line:
        word_list.append(word)
        
for i in range(8 , len(word_list) - 1):    
    description.append(word_list[i])

""" Collect the yearly_revenue, yearly net profit and asking price"""
num = []
tags = soup('dd')
for tag in tags:
    num.append(tag.contents[0])

Yearly_revenue = []
Yearly_net_profit = []
Asking_price = []

for i in range(0, (len(num)- 1)):
    if i % 3 == 0:
        Yearly_revenue.append(num[i])
    elif i % 3 == 1:
        Yearly_net_profit.append(num[i])
    elif i % 3 == 2:
        Asking_price.append(num[i].strip())
        
df = pd.DataFrame({'Yearly_revenue' : Yearly_revenue , 
                   'Yearly_net_profit': Yearly_net_profit , 
                   'Asking_price': Asking_price,
                   'Description' : description})
```

```python
description = []
Industry = []
Category = []
Summary = []
for i in range(8, len(word_list) - 1):
    description.append(word_list[i])
    subsection1 = word_list[i].split('-')
    
    if len(subsection1) == 1:
        subsection2 = subsection1[0].split('–')
        if len(subsection2) == 1:
            Industry.append('')
            Category.append('')
            Summary.append(subsection2[0])
        if len(subsection2) != 1:
            Industry.append(subsection2[0])
            Category.append(subsection2[1])
            Summary.append(subsection2[2])
    if len(subsection1) == 2:
        Industry.append('')
        Category.append(subsection1[0])
        Summary.append(subsection1[1])
    if len(subsection1) == 3:
        Industry.append(subsection1[0])
        Category.append(subsection1[1])
        Summary.append(subsection1[2])
    if len(subsection1) == 4:
        if  (re.search('E-Commerce',word_list[i]) or re.search('E-commerce',word_list[i])) :
            Industry.append('E-Commerce')
            Category.append(subsection1[2])
            Summary.append(subsection1[3])
        else:
            Industry.append(subsection1[0])
            Category.append(subsection1[1])
            Summary.append('')
    if len(subsection1) == 5:
        Industry.append('E-Commerce')
        Category.append(subsection1[3])
        Summary.append(subsection1[4])
    if len(subsection1) == 6:
        Industry.append('E-Commerce')
        Category.append(subsection1[4])
        Summary.append(subsection1[5])
```

```python
df = pd.DataFrame({'Yearly_revenue' : Yearly_revenue , 
                   'Yearly_net_profit': Yearly_net_profit , 
                   'Asking_price': Asking_price,
                   'Description' : description,
                   'Industry': Industry ,
                   'Category': Category,
                   'Summary': Summary})
display(df)
df.to_csv('D:\SFSU\On-campus job\Data Analyst\Sureswift\selling_company_feinternational_1.csv', header = True, index = True, mode = 'w')
```
